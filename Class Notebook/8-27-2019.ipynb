## Second Question

We might ask, how does a player's height or weight depend on the position they play? 

```python
set(major.Position)
```

```python
pos_dict = {x:'Pitcher' for x in list(set(major.Position)) 
            if 'Pitcher' in x}
pos_dict
```

```python
pos_dict2 = {x:'Not Pitcher' for x in list(set(major.Position))
            if not 'Pitcher' in x}
pos_dict2
```

```python
pos_dict.update(pos_dict2)
major_2 = major.copy()

major_2.Position = major_2.Position.map(pos_dict)

major_2
```
